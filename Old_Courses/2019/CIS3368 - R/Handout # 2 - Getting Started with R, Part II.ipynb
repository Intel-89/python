## Case-Scenario 1

```r
# Grades so far
grades = c(71,69,79)         # Creating a vector with all the values for the test.
# Average quiz grade wanted
target_mean_grade = 70       # Creating a variable with the target mean grade.
# Number of quizzes
#quizzes = 4                 # Creating a variable with number of quizzes.
quizzes=length(grades)+1     # Creating a variable with number of quizzes little more efficient, getting 
                             # the lenght of our original vector and adding 1 for the final exam we need
                             # to know the grade in order to get our min target mean
# Needed grade on quiz 4
x4 = quizzes*target_mean_grade - sum (grades) # Calculating our need exam grade in order to get
                                            # the min mean score.
message("Solution")        # Posting a title on our output
cat("The minimun grade Kate needs to have an mean score of 70 is: ", x4)
```

```r
# Final Grades
final_grades = c(grades,61)                # Creating a vector adding the last exam score.
# Calc. Mean for final_grades
kate_mean_grade = mean(final_grades)       # Calculating the mean from all scores.
#kate_mean_grade
# Calc. Standard Deviation for final_grades
kate_sd_grade = sd(final_grades)           # Calculating the standard deviation from all scores.
#kate_sd_grade
# Calc. Min for final_grades
kate_min_grade = min(final_grades)         # Calculating the minimum from all scores.
#kate_min_grade
# Calc. Max for final_grades
kate_max_grade = max(final_grades)         # Calculating the maximum from all scores.
#kate_max_grade
# Calc. 25% for final_grades
kate_1q_grade = quantile(final_grades, 1/4)# Calculating the 1st quantile from all scores.
#kate_1q_grade
# Calc. 75% for final_grades
kate_3q_grade = quantile(final_grades, 3/4)# Calculating the 3rd quantile from all scores.
#kate_3q_grade
# Calc. INterquantile range for final_grades using IQR funtion
kate_iq_grade = IQR(final_grades)
#kate_3q_grade
message("Descriptive Statistics")        # Posting a title on our output
# Summary funtion can also be used
summary(final_grades)                   # Using summary() funtion to get all the statistics value above in a single command, more efficient and elegant.
```

```r
#Frank's Grades
Frank_grades = c(41,69,63,94,99) # Original grades
Frank_target_mean_grade = 70     # Target mean
quizzes=length(Frank_grades)+1   # Number of quizzes
x6 = quizzes*Frank_target_mean_grade - sum (Frank_grades) # Targeted score for last test
F_grades = c(grades,61)          # Adding last grade to the original vector of grades
message("Descriptive Statistics")        # Posting a title on our output
summary(F_grades)                # Descriptive Statistic about the grades' vector
message("Solution")        # Posting a title on our output
cat("The minimun grade Frank needs to have an mean score of 70 is: ", x6)
```

## Case-Scenario 2

```r
n1 = 10               # Creating a variable with qty of men.
n2 = 4                # Creating a variable with qty of women.
y1 = 72000            # Creating a sum of salaries for men.
y2 = 84000            # Creating a sum of salaries for women.

# Doing the actual calculation of the mean salary for all people.
mean_salary=round((n1*y1+n2*y2)/(n1+n2),2) # since result had to many decimal positions it was rounded to common 2 decimals 
#mean_salary
message("Solution")        # Posting a title on our output
cat("The mean salary of all 16 individual is: ", mean_salary)
```
