1. Which of your variables is the response variable?

The response variable is the focus of a question in a study or experiment, is a variable whose value depends upon independent variables. 

2. Which are the explanatory variables?

An explanatory variable is a type of independent variable. The two terms are often used interchangeably. But there is a subtle difference between the two. When a variable is independent, it is not affected at all by any other variables. When a variable isn’t independent for certain, it’s an explanatory variable. They are called explanatory variables because they may offer some explanation

```r
# "file_Handout_3.csv" is a comma separated file with 3 vectors.
myTable = read.table('file_Handout_3.csv', header = TRUE, sep = ',') # Reading the file and creating a dataframe with the data.
message("Garden A variance is:")        # Posting a title on our output.
round(var(myTable$Garden_A),2)          # Using 'var()' to calculate variance from 1st feature and also rounding to 2 decimals places.
message("Garden B variance is:")        # Posting a title on our output.
round(var(myTable$Garden_B),2)          # Using 'var()' to calculate variance from 2nd feature and also rounding to 2 decimals places
message("Garden C variance is:")        # Posting a title on our output. 
round(var(myTable$Garden_C),2)          # Using 'var()' to calculate variance from 3rd feature and also rounding to 2 decimals places.
```

```r
message("Garden A mean is:")        # Posting a title on our output  
round(mean(myTable$Garden_A),2)     # Using 'mean()' to calculate variance from 1st feature and also rounding to 2 decimals places.
message("Garden B mean is:")        # Posting a title on our output  
round(mean(myTable$Garden_B),2)     # Using 'mean()' to calculate variance from 2nd feature and also rounding to 2 decimals places.
message("Garden C mean is:")        # Posting a title on our output  
round(mean(myTable$Garden_C),2)     # Using 'mean()' to calculate variance from 3rd feature and also rounding to 2 decimals places.
```

## Is it a bad idea comparing means when the variances are different? Explain.

### Variances are based on  the expectation of the squared deviation of a random variable from its mean, so from samples with different means you might have same variances; but different varianes might have the same mean. I dont think is a bad idea doing the means comparison even when the variances are different.


```r
ratio1_CB=round(var(myTable$Garden_C)/var(myTable$Garden_B),2)
aBool=ratio1_CB>var(myTable$Garden_B)*4

message('Ratio of variances "C/B" (',ratio1_CB,') is greater than 4 times the variance of B (',round(var(myTable$Garden_B),2),')?: ',aBool)
```
