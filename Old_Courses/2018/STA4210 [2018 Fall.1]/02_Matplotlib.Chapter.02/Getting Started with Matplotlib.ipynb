```python
d=[11,12,13,14,15,16,17]
t0=[15.3,15.4,12.6,12.7,13.2,12.3,11.4]
t1=[26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2=[22.3,20.6,19.8,21.6,21.3,19.4,21.4]


plt.plot(d,t0, label="Tokyo")
plt.plot(d,t1, label="Hawaii")
plt.plot(d,t2, label="Hong Kong")

plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12, fontweight='semibold')
plt.ylabel('Temperature (C)',size=12, fontweight='semibold')

plt.grid(True, linewidth=0.5, color='#aaaaaa', linestyle='-')
plt.title("Daily Temperature of 3 cities in the 2nd week of December",size=14, fontweight='bold')
plt.legend(loc=(0))
```

## Scatter plot

```python
# seed the random number generator to keep results reprodible
np.random.seed(123)

# Generate 10 random numbers around 2 as x-coordinates of the first data series
x0=np.random.rand(10)+1.5

np.random.seed(456)
# Generate 10 random numbers around 3 as y-coordinates of the first data series
y0=np.random.rand(10)+2

np.random.seed(789)

# Generate 10 random numbers around 2 as x-coordinates of the second data series
x1=np.random.rand(10)+2.5

np.random.seed(555)
# Generate 10 random numbers around 3 as y-coordinates of the second data series
y1=np.random.rand(10)+2

plt.scatter(x0,y0)
plt.scatter(x1,y1)
```

## Overlaying multiple data series in a plot

```python
np.random.seed(100)
x=list(range(10))
y=x+np.random.rand(10)-.5
fit=np.polyfit(x,y,1)
print(fit)
yfit=[n*fit[0] for n in x]+fit[1]
plt.scatter(x,y)
plt.plot(yfit,'black')
```

```python
crop=pd.read_csv('crop.txt',delimiter='\t')
crop
```

```python
# Import dataset 
years=crop[crop['Crop']=='SOYBEAN']['Year']
rice=crop[crop['Crop']=='RICE']['Value']
wheat=crop[crop['Crop']=='WHEAT']['Value']
maize=crop[crop['Crop']=='MAIZE']['Value']
soybean=crop[crop['Crop']=='SOYBEAN']['Value']

#Plot the data series

plt.plot(years, rice, label='Rice')
plt.plot(years, wheat, label='Wheat')
plt.plot(years, maize, label='Maize')
plt.plot(years, soybean, label='Soybean')

#Label the x- and y-axes
plt.xlabel('Year',size=12, fontweight='semibold')
plt.ylabel('Thousand Tonnes',size=12, fontweight='semibold')

# Add grid
plt.grid(linestyle='--', color='#888888', alpha=0.4)

# Add The title & legend
plt.title('Total OECD crop production in 1995-2016', size=14, fontweight='semibold')
plt.legend()

# Show the figure
plt.show()
```
